# T-Test

## There is a relationship between BMI and smoking status
The BMI levels of people are different in smokers and non-smokers

Null Hypothesis: There is no relationship between BMI and smoking status

```python
# # Split the data into two groups: BMI_obese and BMI_nonobese
BMI_obese = df[df['BMI'] >= 30]['BMI']
BMI_obese
```

```python
BMI_obese.describe()
```

```python
BMI_nonobese = df[df['BMI'] < 30]['BMI']
BMI_nonobese
```

```python
BMI_nonobese.describe()
```

```python
# Separate the obese people in relation to smoking 100 cigarettes in a lifetime
BMI_obese_data = df[df['BMI'] >= 30]['Smoker']
BMI_obese_data
```

```python
# Separate the non obese people in relation to smoking 100 cigarettes in a lifetime
BMI_nonobese_data = df[df['BMI'] < 30]['Smoker']
BMI_nonobese_data
```

## The T-value of 9 indicates that there is a large difference between the two data sets in regards to smoking.

The small P value indicates that the null hypothesis should be rejected.

```python
t_stat, p_val = ttest_ind(BMI_obese_data, BMI_nonobese_data, equal_var=False)  # Assuming unequal variances with Welch's t-test # noqa

print(f"T-statistic: {t_stat}")
print(f"P-value: {p_val}")
```
